```python
import pandas as pd
```

```python
def ingest_worldbank_data(filename):
    worldbank_df = pd.read_csv(filename, skiprows=4)
    years_df = worldbank_df.set_index(['Country Name']).T
    return years_df, worldbank_df
```

```python
years_df, countries_df = ingest_worldbank_data('/content/API_19_DS2_en_csv_v2_6183479.csv')
```

```python
conditions = ((countries_df['Country Name'].isin(['United States', 'United Kingdom'])) & (countries_df['Indicator Name'].isin(['Urban population (% of total population)',
                                                                                                                               'Urban population growth (annual %)',
                                                                                                                               'Population growth (annual %)', 'Population, total',
                                                                                                                               'Urban population', 'CO2 emissions from solid fuel consumption (kt)',
                                                                                                                               'CO2 emissions from liquid fuel consumption (kt)'])))

population_df = countries_df[conditions]
```

```python
population_df.head(10)
```

```python
mean_values_list = []

for index, row in population_df.iterrows():
    mean_values_list.append(row.iloc[4:].mean() if not row.iloc[4:].isnull().all() else row.iloc[4])
```

```python
for index, mean_value in zip(population_df.index, mean_values_list):
    mask = population_df.index == index
    population_df.loc[mask] = population_df.loc[mask].fillna(mean_value)
```

```python
import matplotlib.pyplot as plt
```

```python
uk_population_data = population_df[(population_df['Country Name'] == 'United Kingdom') & (population_df['Indicator Name'] == 'Population, total')].iloc[:, 4:-1]
plt.figure(figsize=(10, 5))
plt.plot(uk_population_data.columns, uk_population_data.values.flatten(), marker='o', label='United Kingdom')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Total Population Over Years for the United Kingdom')
plt.legend(title='Country Name')
plt.show()
```
